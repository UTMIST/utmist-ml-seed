## 5. Check outputs

````python
import json
from pathlib import Path

# Find the latest Hydra output
output_dirs = sorted(Path("outputs").rglob("metrics.json"))
if output_dirs:
    latest = output_dirs[-1]
    metrics = json.loads(latest.read_text())
    for m in metrics:
        print(f"Epoch {m['epoch']}: train_loss={m['train_loss']:.4f}, val_acc={m['val_acc']:.4f}")
else:
    print("No metrics found yet. Run training first.")
````
